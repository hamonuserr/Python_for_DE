```python
prev_city = None
city_count = 1
day_temp_sum = 0
night_temp_sum = 0


with open('city_weather_avg.csv', 'w', encoding='utf-8') as city_w_add:
    with open('city_we_sorted.csv', 'r', encoding='utf-8') as city_w:
        for i, line in enumerate(city_w):
            num, city, date, day_t, night_t = line.strip().split(',')

            if i == 0:
                continue              
                
            if prev_city:
                if prev_city == city:
                    city_count += 1
                    day_temp_sum += float(day_t)
                    night_temp_sum += float(night_t)
                else:
                    print(f"{prev_city}, avg.temp: {round(((day_temp_sum+night_temp_sum)/city_count), 2)}")
                    city_count = 1
                    day_temp_sum = 0
                    night_temp_sum = 0
                    prev_city = city
            else:
                prev_city = city
                city_count = 1
                day_temp_sum = 0
                night_temp_sum = 0
                
        print(f"{prev_city}, avg.temp: {round(((day_temp_sum+night_temp_sum)/city_count), 2)}")
```
